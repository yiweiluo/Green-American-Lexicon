```python
# Load the saved frames (i.e., tokens within 2 steps of anchor term)
frames = load_frames_for_speech(114)
```

```python
frames.head(10)
```

```python
# Adjectival modifiers of 'alien'
alien_frames = frames[(frames.first_anchor == 'alien') & (frames.edge_str == 'amod-dep')]
alien_frames = alien_frames.drop_duplicates(subset=['speech_id', 'section_id', 'sent_id', 'tok_id'])
print(len(alien_frames))
alien_frames.head()
```

```python
# Adjectival modifiers of 'immigrant'
immigrant_frames = frames[(frames.first_anchor == 'immigrant') & (frames.edge_str == 'amod-dep')]
immigrant_frames = immigrant_frames.drop_duplicates(subset=['speech_id', 'section_id', 'sent_id', 'tok_id'])
print(len(immigrant_frames))
immigrant_frames.head()
```

```python
counter1 = Counter(alien_frames.lemma.values)
counter2 = Counter(immigrant_frames.lemma.values)
prior_counter = counter1 + counter2
ratios = compare_word_counts_with_log_odds(counter1, counter2, prior_counter, prior_weight=100)
sig = ratios[ratios.z_score > 1.96]
print('%d words (%.1f%%) are significantly associated with \'alien\'' % (len(sig), 100 * len(sig) / len(ratios)))
sig
```

```python
ratios = ratios.sort_values(by='log_odds_ratio')  # most negative on top
sig = ratios[ratios.z_score < -1.96]
print('%d words (%.1f%%) are significantly associated with \'immigrant\'' % (len(sig), 100 * len(sig) / len(ratios)))
sig
```
